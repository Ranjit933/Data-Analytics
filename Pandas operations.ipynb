```python
import matplotlib.pyplot as plt
import seaborn as sns

data={
    'Name':['John','Jane','Steve','Linda'],
    'Age':[25,35,None,28],
    'Salary':[50000,60000,45000,None],
    'Dept':['Mktng','Sales','Mktng','Engg'],
    'Gender':['M','F','M','F'],
    'City':['NYC','Chicago','LA','Chicago']    
}

df=pd.DataFrame(data)
df

#Desrtravtive statistics
summary_stats=df.describe()
summary_stats

#Data visualization
sns.set(style='darkgrid')

# Bar plot

sns.barplot(x='Dept',y='Salary',data=df)
plt.title('Average salary by gender')
plt.show()

#  Data preprocessing
df['Dept']=df['Dept'].str.upper()
df

#Data transformation

df['Salary in months']=df['Salary']/12
df

# Data visulaization using Matplotlib

plt.scatter(df['Age'],df['Salary'], color='blue')
plt.xlabel('Age')
plt.ylabel('Salary')
plt.title('Age VS Salary')
plt.show()
```
