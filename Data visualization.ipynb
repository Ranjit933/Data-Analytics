Differents kinds of plot

Lines Plots

```python
import pandas as pd
import matplotlib.pyplot as plt

# create a dataframe

data={'year':[2010,2011,2012,2013,2014,2015],
     'sales':[100,150,120,180,200,250]}

df=pd.DataFrame(data)
df

# create a line plot

plt.plot(df['year'], df['sales'])
plt.xlabel('Year')
plt.ylabel('Sales')
plt.title('Sales over the year')

# Display the plot

plt.show()
```

```python
# Multiple lines

data={'year':[2010,2011,2012,2013,2014,2015],
     'sales1':[100,150,120,180,200,250],
     'sales2':[80,120,110,160,190,220]}

df=pd.DataFrame(data)
df


# create a line plot

plt.plot(df['year'], df['sales1'], label='sales1')
plt.plot(df['year'], df['sales2'], label='sales2')
plt.xlabel('Year')
plt.ylabel('Sales')
plt.title('Sales over the year')
plt.legend()         

# Display the plot

plt.show()
```

Bar plot


```python
data={'year':[2010,2011,2012,2013,2014,2015],
     'sales':[100,150,120,180,200,250]}

df=pd.DataFrame(data)
df

# Create a bar plot


plt.bar(df['year'], df['sales'])
plt.xlabel('Year')
plt.xlabel('Year')
plt.ylabel('Sales')
plt.title('Sales by year')

# show the plot

plt.show()
```

Scatter plots

```python
data={'x':[1,2,3,4,5],
     'y':[2,4,6,8,10]
     }

df=pd.DataFrame(data)
df


plt.scatter(df['x'], df['y'])
plt.xlabel('x')
plt.xlabel('y')
plt.ylabel('SCATTER')

# show the plot

plt.show()
```

```python
# With a Trend line
import seaborn as sns

# create a scatterplot with Trendline
sns.regplot(x='x', y='y', data=df)
plt.xlabel('x')
plt.xlabel('y')
plt.ylabel('SCATTER PLOT WITH TREND LINE')

# show the plot

plt.show()
```

Histogram

```python
import numpy as np

# Generate a randam data

np.random.seed(42)
data=np.random.normal(loc=50, scale=10, size=1000)

# create a DataFrame

df=pd.DataFrame({'values':data})

#create a histogram

plt.hist(df['values'], bins=20, color='skyblue', edgecolor='Black')
plt.xlabel('values')
plt.xlabel('Frequency')
plt.title('HISTOGRAM')

# show the plot

plt.show()
```

BOX PLOTS

```python
data={'category':['A','A','A','B','B','B','B'],
       'values':[40,15,20,25,30,35,40]}

df=pd.DataFrame(data)
df

sns.boxplot(x='category', y='values', data=df)
plt.xlabel('Category')
plt.xlabel('values')
plt.title('BOX PLOT')

# show the plot

plt.show()
```

Heatmaps

```python
data={'A':[1,2,3],
      'B':[4,5,6],
       'c':[7,8,9]}

df=pd.DataFrame(data)
df


sns.heatmap(df, annot=True, cmap='YlGnBu')
plt.xlabel('ColumnS')
plt.xlabel('Rows')
plt.title('HEATMAP')

# show the plot

plt.show()
```
